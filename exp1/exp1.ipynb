Task 1.15 & 1.16

```python

merge_df1 = pd.DataFrame(
	{
		"id": [1, 2, 3, 4],
		"A": ["A1", "A2", "A3", "A4"],
		"B": ["B1", "B2", "B3", "B4"],
	}
)
merge_df2 = pd.DataFrame(
	{
		"id": [3, 4, 5, 6],
		"A": ["A3", "A4", "A5", "A6"],
		"colE": ["E3", "E4", "E5", "E6"],
	}
)

print("Inner join")
print(pd.merge(merge_df1, merge_df2, on="id"))

print("\nLeft join")
print(merge_df1.merge(merge_df2, on="id", how="left"))

print("\nRight join")
print(pd.merge(merge_df1, merge_df2, on="id", how="right"))

print("\nFull outer join")
print(pd.merge(merge_df1, merge_df2, on="id", how="outer"))


```

```python
print("\nLeft anti-join")
left_merge = merge_df1.merge(merge_df2, on="A", how="left", indicator=True)
left_only_keys = left_merge.loc[left_merge["_merge"] == "left_only", "A"]
print(merge_df1[merge_df1["A"].isin(left_only_keys)])

print("\nRight anti-join")
right_merge = merge_df1.merge(merge_df2, on="A", how="right", indicator=True)
right_only_keys = right_merge.loc[right_merge["_merge"] == "right_only", "A"]
print(merge_df2[merge_df2["A"].isin(right_only_keys)])

print("\nFull anti-join")
full_merge = merge_df1.merge(merge_df2, on="A", how="outer", indicator=True)
both_keys = full_merge.loc[full_merge["_merge"] == "both", "A"]
print(full_merge[~full_merge["A"].isin(both_keys)])

print("\nMerge on different column names")
diff_left = pd.DataFrame({"A": [1, 2, 3], "X": [10, 20, 30]})
diff_right = pd.DataFrame({"B": [2, 3, 4], "Y": [200, 300, 400]})
print(diff_left.merge(diff_right, left_on="A", right_on="B", how="inner"))

```

```python
print("\nMerge on multiple columns")
multi_left = pd.DataFrame({"A": [1, 1, 2], "B": ["x", "y", "z"], "L": [10, 20, 30]})
multi_right = pd.DataFrame({"A": [1, 2, 2], "B": ["x", "z", "q"], "R": [100, 200, 300]})
print(multi_left.merge(multi_right, on=["A", "B"], how="inner"))


print("Task 1.16")
task16_df = merge_df1.merge(merge_df2[["id", "colE"]], on="id", how="left")
print(task16_df)
```
